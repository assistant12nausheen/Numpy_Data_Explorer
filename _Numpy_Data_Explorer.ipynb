```python
import numpy as np
```

```python
data = np.array([[10, 20, 30],
                 [40, 50, 60],
                 [70, 80, 90]])
```

```python
# Basic math
print("Sum:", np.sum(data))
print("Mean:", np.mean(data))
print("Max:", np.max(data))
print("Min:", np.min(data))
print("Std Dev:", np.std(data))
```

```python
# Axis-wise (axis=0 → column-wise, axis=1 → row-wise)
print("Column sums (axis=0):", np.sum(data, axis=0))
print("Row means (axis=1):", np.mean(data, axis=1))
```

```python
# Math operations
arr = np.array([1, 4, 9, 16, 25])
print("Square root:", np.sqrt(arr))
print("Multiply by 2:", arr * 2)
print("Add 10:", arr + 10)
```

```python
# RESHAPING — change shape without changing data
arr = np.arange(12)  
print("Original: ", arr)

reshaped = arr.reshape(3, 4)   
print("Reshaped to 3x4: ", reshaped)

reshaped2 = arr.reshape(2, 2, 3)  
print("Reshaped to 2x2x3: ", reshaped2)
```

```python
# Flatten back to 1D
flat = reshaped2.flatten()
print("Flattened: ", flat)
```

```python
# BROADCASTING — operate on different shaped arrays
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
row_vector = np.array([10, 20, 30])
```

```python
# NumPy broadcasts row_vector across each row automatically!
result = matrix + row_vector
print("Matrix + row vector: ", result)
```

```python
# Create sample data
sales_data = np.array([[100, 200, 300],
                        [150, 250, 350],
                        [120, 220, 320]])
```

```python
# Save as .npy (single array)
np.save('sales_data.npy', sales_data)
print("Saved sales_data.npy")
```

```python
# Load it back
loaded = np.load('sales_data.npy')
print("Loaded array: ", loaded)
```

```python
# Save multiple arrays in one .npz file
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
np.savez('both_arrays.npz', x_data=x, y_data=y)
```

```python
# Load .npz
data = np.load('both_arrays.npz')
print("x_data:", data['x_data'])
print("y_data:", data['y_data'])
```

```python
# Save as CSV (human-readable)
np.savetxt('sales.csv', sales_data, delimiter=',', fmt='%d')
loaded_csv = np.loadtxt('sales.csv', delimiter=',')
print("Loaded from CSV: ", loaded_csv)
```

```python
import time
```

```python
SIZE = 1_000_000  
```

```python
# Python list approach
py_list = list(range(SIZE))
start = time.time()
result_list = [x * 2 for x in py_list]
list_time = time.time() - start

# NumPy array approach
np_arr = np.arange(SIZE)
start = time.time()
result_numpy = np_arr * 2
numpy_time = time.time() - start
```

```python
print(f"Python list time : {list_time:.4f} seconds")
print(f"NumPy array time : {numpy_time:.4f} seconds")
print(f"NumPy is {list_time/numpy_time:.1f}x faster!")
```

```python
# Memory comparison
import sys
print(f"List memory  : {sys.getsizeof(py_list):,} bytes")
print(f"NumPy memory : {np_arr.nbytes:,} bytes")
```
